```python
import urllib.request as req
from urllib.error import URLError, HTTPError

url_list = ['http://post.phinf.naver.net/20160621_169/1466482468068lmSHj_JPEG/If7GeIbOPZuYwI-GI3xU7ENRrlfI.jpg',
            'http://www.google.com']

path_list = ['c:/my_web/myexam/test4.jpg', 'c:/my_web/myexam/index4.jpg']

for i, url in enumerate(url_list) :
    try :
        response = req.urlopen(url)
        contents = response.read()
        
        print('Header - Info {} : {}'.format(i, response.info()))
        print('Http Status Code {}'.format(response.getcode()))
        
        # 저장경로에 불러온 내용 담기
        with open(path_list[i], 'wb') as c :
            c.write(contents)
            
    except URLError as e:
        print('Download Failed')
        print(e)
        
    except HTTPError as e:
        print('Download Failed')
        print(e)
        
    else :
        print("Download Suceed")
```

```python
import urllib.request as req
from urllib.error import URLError, HTTPError

url_list = ['http://post.phinf.naver.net/20160621_169/1466482468068lmSHj_JPEG/If7GeIbOPZuYwI-GI3xU7ENRrlfI.jpg',
            'http://www.google.com']

path_list = ['c:/my_web/myexam/test4.jpg', 'c:/my_web/myexam/index4.jpg']

for i, url in enumerate(url_list) :
    try :
        response = req.urlopen(url)
        contents = response.read()
        
        # 중간 정보 확인
        print("Header - info {} : {}".format(i, response.info()))
        print("HTTP Status Code : {}".format(response.getcode()))
        
        # 저장 경로에 받아온 내용 집어넣기 
        with open(path_list[i], 'wb') as c :
            c.write(contents)
            
    except HTTPError as e :
        print("Download Failed")
        print('HTTPError Code :', e.code)
        
    except URLError as e:
        print('Download Failed')
        print("URLError Reson :", e.reason)

    else :
        print("Download Succeed")        
```
